```python
# Load the Whisper model
model = whisper.load_model("small")  # You can change to small, large, etc.

# Transcribe the audio file, something is wrong with the audio file 023_06
result = model.transcribe("/Users/yueyan/Downloads/IW_023_06_TL.WAV", word_timestamps = True)  # Replace with your audio file path
```

```python
# Load the Whisper model
model = whisper.load_model("medium")

# Transcribe the audio
result = model.transcribe("/Users/yueyan/Documents/project/wearable/media/043_08/IW_043_08_TL.WAV")
```

```python
# Function to convert seconds to minute:seconds format
def format_timestamp(seconds):
    minutes = int(seconds // 60)
    seconds = int(seconds % 60)
    return f"{minutes:02}:{seconds:02}"

# Prepare the transcription in the desired format
transcription_output = []
for segment in result['segments']:
    entry = {
        "Timestamp": f"{format_timestamp(segment['start'])} - {format_timestamp(segment['end'])}",
        "Text": segment['text'],
        "Lang": "en",  # Modify according to the language detected
        "Speaker": "Unknown"  # Modify this if speaker identification is needed or available
    }
    transcription_output.append(entry)

# Define the directory and file name for saving
output_dir = "./Documents/project/wearable/transcription"  # Set your desired directory path here
output_filename = "043_08_transcription_medium.json"

# Ensure the directory exists
os.makedirs(output_dir, exist_ok=True)

# Save the transcription to a JSON file in the specified directory
output_file_path = os.path.join(output_dir, output_filename)
with open(output_file_path, "w", encoding="utf-8") as outfile:
    json.dump(transcription_output, outfile, ensure_ascii=False, indent=4)

print(f"Transcription saved as {output_file_path}")
```
